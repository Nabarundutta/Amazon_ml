```python
import pandas as pd
import numpy as np
```

```python
import gensim
```

```python
df = pd.read_csv('processed_imdb.csv')
df.head()
```

```python
df.drop(columns=['review','tokens'],axis=1,inplace=True)
```

```python
df.head()
```

```python
import gensim
from gensim.models import Word2Vec
```

```python
word2vec_model = Word2Vec(sentences=df['stemmed_tokens'],vector_size=100,window=5,min_count=2)
```

```python
def tokens_to_vectors(tokens, model):
    return [model.wv[word] for word in tokens if word in model.wv]
```

```python
df['word_vectors'] = df['stemmed_tokens'].apply(lambda x: tokens_to_vectors(x, word2vec_model))
```
